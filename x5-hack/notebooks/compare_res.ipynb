```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('.././data/subm_pred_92.csv', sep=';')
df2 = pd.read_csv('.././data/subm_pred_91.csv', sep=';')
df1.head()
```

```python
rows = []
for i, row in df1.iterrows():
    if df1.loc[i, 'annotation'] != df2.loc[i, 'annotation']:
        rows.append({
            'sample': df1.loc[i, 'sample'],
            'annotation 0.92': df1.loc[i, 'annotation'],
            'annotation 0.91': df2.loc[i, 'annotation'],
        })

df = pd.DataFrame(rows)
df.to_csv('annotations.csv', index=False)
```

```python
for i, row in df1.iterrows():
    was_b_type = False
    was_b_brand = False
    for span in eval(row['annotation']):
        if not was_b_type and 'B-TYPE' in span:
            was_b_type = True
            continue
        if not was_b_brand and 'B-BRAND' in span:
            was_b_brand = True
            continue
        if 'B-TYPE' in span and  was_b_type or 'B-BRAND' in span and was_b_brand:
            print(f"{row['sample']}: {row['annotation']}")
            
```
